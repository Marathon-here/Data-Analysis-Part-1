```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("AB_NYC_2019.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.dropna(inplace = True)
df.drop_duplicates(inplace = True)
```

```python
df.describe()
```

```python
#we dont need id and hostid , latitude , longitude calculations so convert them to strings
# see that if describe fun calculations as negatives means we don't need that type of column calculations
df["id"] = df["id"].astype('string')
df["host_id"] = df["host_id"].astype('string')
df["latitude"] = df["latitude"].astype('string')
df["longitude"] = df["longitude"].astype('string')
```

```python
df.describe()
```
