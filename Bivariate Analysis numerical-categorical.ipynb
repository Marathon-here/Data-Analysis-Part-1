```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset("titanic")
```

```python
df.head()
```

# BAR plot 
(x = category, y= numerical)

```python
sns.barplot(x = 'class',y='age',data=df)  #bar garph based on avg age estimator of each category
```

```python
sns.barplot(x = 'class',y='fare', data=df , estimator = 'sum') # estimated based on sum
```

```python
sns.barplot(x = 'class',y='fare',hue ='sex' , data=df , estimator = 'sum')
```

## Box Plot

### Box Plot: Understanding Distribution and Outliers

```python
sns.boxplot(x = 'class',y='fare',data=df )
```

```python
sns.boxplot(x = 'class',y='age',hue ='survived' , data=df )
```

## VIOLIN PLOT 

Violin Plot: Combining Distribution and Box Plot
A violin plot combines the benefits of a box plot and a kernel density plot. It shows the distribution of data along with its probability density, making it a great tool for understanding the distribution across different categories.

### Blackthick border is interquatile range and white dot  is median value 
### simple line of violin goes to max and min value of data


```python
sns.violinplot(x = 'class',y='age',data=df )
```

```python
sns.violinplot(x = 'class',y='fare',data=df )
```

## DistPlot
### Distribution Plot: Comparing Categories

```python
sns.distplot(df[df['class']=="Second"]['age'])
```

```python
sns.distplot(df[df["survived"] == 0]["age"], hist=False)
sns.distplot(df[df["survived"] == 1]["age"], hist=False)
```
