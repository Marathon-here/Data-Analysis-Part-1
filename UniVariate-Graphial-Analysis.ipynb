```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Univariate Analysis - CATEGORICAL DATA

```python
df = pd.read_csv("AB_NYC_2019.csv")
```

```python
df.columns.tolist()
```

```python
df.dtypes
```

## Bar Chat

```python
sns.countplot(data = df, x = "neighbourhood_group")
```

```python
sns.countplot(data = df, x = "room_type")
```

```python
sns.countplot(data = df, x = "neighbourhood_group", hue = 'room_type')
```

## pie Chart  
### we use the pie to see percentage or composition of a perticular category

```python
df['room_type'].value_counts().plot(kind = "pie")
```

```python
df['room_type'].value_counts().plot(kind = "pie",autopct = "%.2f")
```

```python
df['neighbourhood_group'].value_counts().plot(kind = "pie",autopct = "%.2f")
```

```python
df['room_type'].value_counts().plot(kind = "bar")
```

# Univariate Analysis on Numerical DATA

## Histogram -count
### mostly use histogram when you have the less data then only, you can clearly visualize else if the data is more means go for distplot

```python
# here we can observe data is more we can't clearly see the data
sns.histplot(data = df,x="price")
```

```python
sns.histplot(data = df[df["price"]<1000],x="price")
```

```python
sns.histplot(data = df,x="availability_365")
```

## DistPlot

### gives probability distribution of data

```python
sns.distplot(df["price"])
```

```python
sns.distplot(df["availability_365"])
```

## box plot
### Box Plot: Identifying Outliers
### A box plot is useful for visualizing the spread of the data, including the median, quartiles, and outliers. It helps us identify if there are extreme values (outliers) within a dataset.


```python
sns.boxplot(df["price"])  #to find outliers
```

```python
sns.boxplot(df["availability_365"])  #to find outliers
```

## violin plot

### instead of box plot we can also use violin plot to see count as well as probability distribution

### Violin Plot: Combining Distribution and Box Plot
### A violin plot combines aspects of both box plots and density plots. It shows the distribution of the data while also highlighting its density, providing a more comprehensive visualization of the distribution.


### Blackthick border is interquatile range and white dot  is median value 
### simple line of violin goes to max and min value of data



```python
sns.violinplot(df["availability_365"])
```

```python
sns.violinplot(df["price"])
```
