#### Filling Missing Values - Imputation
#### numerical data - use mean¶

```python
df.columns
```

```python
df3 = df.copy()
```

```python
df3.head()
```

```python
#fill the null values with mean value of data
mean_value = df3["reviews_per_month"].mean()
df3["reviews_per_month"].fillna(mean_value,inplace = True)
# df3.fillna({"reviews_per_month": mean_value}, inplace=True)
```

```python
df3.head()
```

```python
df3.isna().sum()
```

### to fill the categorical data values

```python
df3["last_review"].fillna(df3["last_review"].value_counts().index[0],inplace=True)
```

```python
df3.isna().sum()
```

### method-2 for categorical data

```python
df4 = df.copy()
```

```python
df4.isna().sum()
```

```python
#filling the missing with new category not_reviewd as value of missing value
df4["last_review"].fillna("Not_Reviewed", inplace=True) 
```

```python
df4.isna().sum()
```
