### Why Handle Missing Values?
#### Missing values can arise due to data collection errors, system failures, or incomplete responses. If left unaddressed, they can lead to biased conclusions and poor model performance.

### Methods for Handling Missing Values
### There are several ways to handle missing values:

### Removing Data
### Removing columns with too many missing values.
### Removing rows that contain missing values.
### Filling Missing Values (Imputation)
#### For numerical data: Use mean, median, or mode.
#### For categorical data: Use mode or introduce a new category.
#### Advanced Imputation Techniques
#### Interpolation to estimate missing values.
#### Predictive models to estimate missing values based on other variables.

### to avoid manual count for each category @2 methods
#### -run for loop for each categorytype to the given column
#### *create a dictory to store each caterogy count
#### *run for loop for each unique category
#### *count the items for each categorytype
#### store in dictonary

```python
categories = {}
for name in df['Category'].unique():
    count = 0
    for i in df['Category']:
        if(i == name):
            count+=1
        categories[name] = count
            
print(categories)
            
```

```python
for name in df['Category'].unique():
    count = 0
    for i in df['Category']:
        if(i == name):
            count+=1
    print(name,":",count)
```
